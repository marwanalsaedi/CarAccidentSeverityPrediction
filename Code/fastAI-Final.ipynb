```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
pd.set_option('display.max_columns', None)
s="all"
data=pd.read_csv(s+".csv")
X=data.copy(True)
```

```python
from datetime import date
XX=X.copy(True)
XX['Start_Time']=pd.to_datetime(XX['Start_Time'])
XX['Year']=XX['Start_Time'].dt.year
XX['Month']=XX['Start_Time'].dt.month
XX['Day']=XX['Start_Time'].dt.day
XX['Hour']=XX['Start_Time'].dt.hour
#XX['DayofYear']=XX['Start_Time'].dt.dayofyear
#XX['WeekofYear']=XX['Start_Time'].dt.weekofyear
XX['Dayofweek']=XX['Start_Time'].dt.dayofweek
#XX['YearMod']=XX['Start_Time'].dt.year%4
XX['Weekend']=XX['Start_Time'].dt.dayofweek>=5
#XX['WeekendFriday']=XX['Start_Time'].dt.dayofweek>=4
```

```python
XX.head()
```

```python
XX.describe(include="all")
```

```python
print(XX.shape)
y=XX['Severity'].copy(True)
y=(y>2)==True
XX['Severity']=y.copy()
XX['Weather_id']=XX['Weather_Condition'].apply(lambda x: x.split("-")[0])
X=XX.drop(columns=['Start_Time','Year','Weather_Condition'])#.dropna()
X.shape
```

```python
from fastai.tabular.all import *
al=X.copy(True)

f = TrainTestSplitter(test_size=0.1, random_state=42)
train,test = f(al)




splits = RandomSplitter(valid_pct=0.1)(range_of(train))

dls = TabularDataLoaders.from_df(al,y_names=["Severity"], cat_names = ['Zipcode', 'Weather_id', 'Month','Day','Hour','Weekend','State'],
    cont_names = ['Start_Lat', 'Start_Lng', 'Temperature','Humidity','Pressure','Wind_Speed'],
    procs = [Categorify,FillMissing, Normalize],y_block=CategoryBlock  ,splits=splits)
```

```python
t=pd.DataFrame(al,index=test)
y_true=t['Severity']
t.drop(columns='Severity',inplace=True)
```

```python
to=dls.dataloaders(bs=2**17)
learn = tabular_learner(dls, metrics=[RocAucBinary(),accuracy],lr=0.003 )
learn.fit_one_cycle(1)
```

```python
learn.fit_one_cycle(1)
```

```python
dl = learn.dls.test_dl(t)
a,b=learn.get_preds(dl=dl)
accuracy(a,Tensor(y_true.values))
from sklearn.metrics import confusion_matrix,classification_report
q=[]
for row in a:
    q.append(row[1]>=0.5)

print(classification_report(y_true,q))
```

```python
learn.show_results()
```

```python
learn.lr=1e-4
learn.fit_one_cycle(1)
```
